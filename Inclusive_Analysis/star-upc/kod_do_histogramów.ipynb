```python
import numpy as np
import matplotlib.pyplot as plt
import os, fnmatch
```

```python
filelist = fnmatch.filter(os.listdir('./star-upc/src/mains'), '*.cxx')
# sortlist = [int(file[file.rfind("_")+1:file.rfind(".")]) for file in nptvlist]
# sortlist, nptvlist = zip(*sorted(zip(sortlist, nptvlist)))
print(filelist)
filelist = ["./star-upc/src/mains/" + file for file in filelist]
```

```python


for file in filelist:
    lines_to_use = []
    print("#"*len(file))
    print(file)
    print("#"*len(file))
    print("")
    with open(file) as f:
        lines = f.readlines()
        for line in lines:
            if line.find("ROOT::TThreadedObject<TH")>=0:
                lines_to_use.append(line)
                
    for line in lines_to_use:
        print("std::shared_ptr" + line[line.find("<"):line.find("=")-1] + "Local;")

    print("")

    for line in lines_to_use:
        print(line[line.find(">")+2:line.find("=")-1] + "Local = " + line[line.find(">")+2:line.find("=")-1] + ".Get();")
        
    print("")
    
    for line in lines_to_use:
        print("std::shared_ptr" + line[line.find("<"):line.find("=")-1] + "Final;")

    print("")

    for line in lines_to_use:
        print(line[line.find(">")+2:line.find("=")-1] + "Final = " + line[line.find(">")+2:line.find("=")-1] + ".Merge();")
        
    print("")

    for line in lines_to_use:
        print(line[line.find(">")+2:line.find("=")-1] + "Final->SetMinimum(0);")
        
    print("")

    print("outputFileHist->cd();")
    for line in lines_to_use:
        print(line[line.find(">")+2:line.find("=")-1] + "Final->Write();")
    print("outputFileHist->Close();")
    
    print("")
```
